```python
import osmnx as ox
import networkx as nx
```

```python
lugar = "Paran√°, Toledo"
g = ox.graph_from_place(lugar, network_type="drive")
```

```python
type(g)
```

```python
ox.plot_graph(g, node_size=1, node_color="blue")
```

```python
origem= list(g.nodes)[100]
destino= list(g.nodes)[-15]

def heuristic(n1, n2):
    # Supondo que n1 e n2 s√£o inteiros, a heur√≠stica pode ser a diferen√ßa absoluta
    return abs(n1 - n2)

# Executa o A* com a heur√≠stica definida
caminho = nx.astar_path(g, origem, destino, heuristic=heuristic)
```

```python
fig, ax = ox.plot_graph_route(g, caminho, route_linewidth=2, node_size=0, bgcolor="w")
```
